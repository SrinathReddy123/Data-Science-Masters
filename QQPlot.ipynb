Check Normal distribution using Quantile Quantile plot

```python
# importing all the necessary libraries

import scipy.stats as stat
import pylab 
import numpy as np

# pylab is used to display some diagrams, figures which we create with the help of matplotlib.
```

```python
# we try to import some dataset from seaborn

import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()

# loading iris dataset into dataframe named df
```

```python
# Now for each feature we try to find the distribution of it is Normal or not.
```

```python
import matplotlib.pyplot as plt

```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data_kde(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
    
    
# here  plot_data() takes 2 parameters, 1 is dataframe and other is feature name for which we want to test if it 
# follows Normal distribution or not.

# For plotting we need matplotlib hence we import the library.

# probplot() takes 2 parameters, feature from dataframe and type of distribution for which we need to check
# and it checks whether feature is of specified distribution or not and then it plots using pylab.

# kde=True it means Kernel density estimator means it will smoothen the curve.
```

probplot():

Calculate quantiles for a probability plot, and optionally show the plot.

Generates a probability plot of sample data against the quantiles of a
specified theoretical distribution (the normal distribution by default).
`probplot` optionally calculates a best-fit line for the data and plots the
results using Matplotlib or a given plot function.

hover mouse onto method -> then click shift+tab - to explain about the description of predefined method.

```python
# checking if sepal_width follows normal distribution or not.

plot_data_kde(df,'sepal_width')
```

If we smoothen the histogram then we get the normal distribution curve.

stat.probplot(df[feature],dist='norm',plot=pylab) -> probability plot curve
The right hand side plot is taking df[feature] as X axis and normal distribution on Y axis, and then it is plotting the probability plot curve

Theotitical quantiles are coming from normal distribution and our dataframe feature is on Y axis.

when most of the points lie on straight line, then we can say that distribution is Normal distribution.


On left side graph from the histogram and kde it looks like Normal distribution.

In QQ Plot, on X axis, we have normal distribution which is initialized and on Y axis we have feature value.

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'sepal_width')
```

```python
# Let's try for other feature(petal width)

plot_data(df,'petal_width')
```

```python
# So we can see petal width is not following Normal distribution.
```

```python
plot_data_kde(df,'petal_width')
```

```python
plot_data(df,'petal_length')
```
