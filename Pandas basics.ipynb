```python
# Dataframe simply means a table(collection of rows and columns)

# let's check the type of the column that is selected
type(df['location_id'])

# It is Series. Series can simply be considered as equivalent to List. In List we don't have indices that are 
# actually visible.But series will give the indexes as well.

# Eitehr we select a row or a column, it will be Series.
```

```python
# converting into a list

list(df['location_id'])
```

```python
# let's try to pass columns as a list.

df[['location_id']]
```

```python
# find the type of above output.

type(df[['location_id']])

# It's dataframe, but not a series.
# So we can do Series based operations and also Dataframe based operations.
```

```python
# selecting multiple columns.
# For 1 column we can pass it as String,but for getting multiple columns we should give it as List.

df['keywords','status']
```

```python
df[['keywords','status']]

# Through a list, we can pass any number of columns that we want, else we get a Key error.
```
