Matplotlib is a popular data visualization library for Python.

It allows users to create a wide range of static, animated, and interactive visualizations in Python.

11. Line plots: Line plots are used to show trends over time or across a continuous variable. They can be created using the plot() function in Pyplot.

2. Scatter plots: Scatter plots are used to show the relationship between two variables. They can be created using the scatter() function in Pyplot.

3. Bar plots: Bar plots are used to compare values across categories. They can be created using the bar() function in Pyplot.

A scatter plot is a type of data visualization that shows the relationship between two variables. 

Each point on the plot represents a single observation, with one variable represented on the x-axis and the other variable represented on the y-axis.

Scatter plots are commonly used to explore the relationship between two quantitative variables and to look for patterns or trends in the data.

To generate data for x and y, you can use the NumPy library in Python. Here is an example code that generates 100 random values for x and y:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

To plot a scatter plot of this data, you can use the scatter() function from the Pyplot module of Matplotlib. Here's an example code:

```python
import matplotlib.pyplot as plt

# Create scatter plot of x and y
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot of X and Y')
plt.show()
```

This code creates a scatter plot of x and y using Matplotlib's scatter() function. The xlabel(), ylabel(), and title() functions are used to add labels and a title to the plot. Finally, the show() function is used to display the plot. 

The resulting scatter plot should show the relationship between the x and y values, with each point representing a single observation.

The subplot() function in Matplotlib is used to create multiple plots within a single figure.

It takes three arguments: the number of rows, the number of columns, and the index of the current plot.

By using the subplot() function, we can create a grid of subplots and specify which subplot to plot in.

```python
Here's an example code that uses the subplot() function to create four line plots:
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots arranged in a 2x2 grid
fig, axs = plt.subplots(2, 2)

# Plot each line plot in a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles and axis labels to each subplot
axs[0, 0].set_title('Line Plot 1')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')

axs[0, 1].set_title('Line Plot 2')
axs[0, 1].set_xlabel('X values')
axs[0, 1].set_ylabel('Y values')

axs[1, 0].set_title('Line Plot 3')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

axs[1, 1].set_title('Line Plot 4')
axs[1, 1].set_xlabel('X values')
axs[1, 1].set_ylabel('Y values')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# Display the figure
plt.show()
```

This code creates a figure with four subplots arranged in a 2x2 grid using the subplots() function. 

It then plots each line plot in a separate subplot using the plot() function and specifies the subplot to plot in using the indexing of the axs array. 

Finally, it adds titles and axis labels to each subplot and adjusts the spacing between subplots using the subplots_adjust() function. The resulting figure should show four separate line plots.

A bar plot, also known as a bar chart, is a plot that uses rectangular bars to represent the data.

It is commonly used to compare the values of different categories or groups.

The height or length of each bar represents the value of the corresponding category or group. 

Bar plots can be used to show changes in data over time, or to compare the values of different categories or groups.

```python
A horizontal bar plot is a bar plot where the bars are arranged horizontally instead of vertically.
```

To plot a bar plot and a horizontal bar plot using the given data, we can use the bar() function and the barh() function in Matplotlib, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure with a bar plot and a horizontal bar plot
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Plot the bar plot
axs[0].bar(company, profit)
axs[0].set_title('Bar Plot')
axs[0].set_xlabel('Company')
axs[0].set_ylabel('Profit')

# Plot the horizontal bar plot
axs[1].barh(company, profit)
axs[1].set_title('Horizontal Bar Plot')
axs[1].set_xlabel('Profit')
axs[1].set_ylabel('Company')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.4)

# Display the figure
plt.show()
```

In this code, we first define the data for the bar plot using two NumPy arrays. We then create a figure with a bar plot and a horizontal bar plot using the subplots() function.

We plot the bar plot using the bar() function and the horizontal bar plot using the barh() function, and specify the company and profit data as the x and y values. 

We then add titles, x-labels, and y-labels to both plots using the set_title(), set_xlabel(), and set_ylabel() functions. Finally, we adjust the spacing between the subplots using the subplots_adjust() function and display the figure using the show() function. 

The resulting figure should show a bar plot and a horizontal bar plot of the profit data for each company.

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that displays the median, quartiles, and extreme values of the data. 

The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The median is shown as a line inside the box.

The whiskers extend from the box to the minimum and maximum values, excluding outliers. Outliers are points that are more than 1.5 times the IQR away from the box.

Box plots are commonly used to identify the distribution of data, as well as to identify any outliers or extreme values in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
```

```python
# Create a figure with a box plot
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot')
ax.set_ylabel('Value')

# Display the figure
plt.show()
```

In this code, we first define the data for the box plot using two NumPy arrays, box1 and box2. We then create a list data that contains both arrays.

We create a figure with a box plot using the subplots() function and specify ax as the axis for the box plot. We plot the box plot using the boxplot() function, and specify the data as the input. 

We then add tick labels, title, and y-label to the plot using the set_xticklabels(), set_title(), and set_ylabel() functions. Finally, we display the figure using the show() function. The resulting figure should show a box plot of the box1 and box2 data.
